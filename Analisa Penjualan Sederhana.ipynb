```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv(r'Data2.csv')
```

```python
data
```

# Mencari Harga diskon dari Harga Satuan

```python
data['Harga Diskon'] =data['Harga Satuan'] -  data['Harga Satuan'] * data['Diskon']
```

```python
data  
```

```python
disc = data.groupby(['Harga Satuan','Harga Diskon'])['Diskon'].value_counts().reset_index().sort_values(by='Diskon', ascending=False)
disc
```

```python
sns.countplot(x='Diskon',data=disc)
plt.title('Distribusi of Diskon')
plt.show()
```

Diskon terbanyak diangka 20 %

# Pengaruh Diskon dengan Pejualan Bersih

```python
diskon = data[['Tanggal','Nama Pelanggan','Nama Produk','Diskon','Harga Diskon','Penjualan Bersih']].sort_values(by = 'Diskon', ascending=False)
diskon
```

```python
sns.barplot(x='Tanggal',y ='Penjualan Bersih', hue='Diskon',data = diskon)
plt.title('Pengaruh Diskon dengan Penjualan Bersih')
plt.show()
```

Penghasilan terbanyak dari semua toko terdapat pada tanggal 2 agustus dengan memberikan diskon 20 %, Namun pada tanggal 1 Agustus Justru dengan hanya diskon 10 % memiliki penghasilan terbanyak dari pada yang diberikan diskon 10 %, Mungkin ada 1 produk yang membuatnya sangat laku 

```python
data[data['Tanggal'] =='3-Aug'].max().reset_index()
```

# Keuntungan Pertoko
